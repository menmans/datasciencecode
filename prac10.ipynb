```python
#List down features and theit types
```

```python
features=iris.columns[:-1] # Exclude the 'lable' column
feature_types=iris[features].dtypes
```

```python
#Identify Outliers
```

```python
outliers={}
for feature in features:
    Q1=iris[feature].quantile(0.25)
    Q3=iris[feature].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    feature_outliers=iris[(iris[feature] < lower_bound) | (iris[feature] > upper_bound)]
```

```python
outliers[feature]=feature_outliers
```

```python
print("outliers:")
for feature,outlier_data in outliers.items():
    print(f"{feature}: {outlier_data.index.tolist()}")
```
