```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("temparature_data.csv")
```

```python
print(data.head())
```

```python
print("Column Names:", data.columns)
```

```python
# Average temperature, dew point, aur wind speed calculate karna
average_temperature = data['Temperature'].mean()
average_dew_point = data['Dew Point'].mean()
average_wind_speed = data['Wind Speed'].mean()

print("Average Temperature:", average_temperature)
print("Average Dew Point:", average_dew_point)
print("Average Wind Speed:", average_wind_speed)
```

```python
# Data visualization
plt.figure(figsize=(10, 6))
```

```python
# Temperature visualization
plt.subplot(3, 1, 1)
plt.hist(data['Temperature'], bins=30, color='blue', alpha=0.7)
plt.axvline(average_temperature, color='red', linestyle='dashed', linewidth=2)
plt.title('Temperature Distribution')
plt.xlabel('Temperature (F)')
plt.ylabel('Frequency')
plt.legend(['Average'])
```

```python
# Dew Point visualization
plt.subplot(3, 1, 2)
plt.hist(data['Dew Point'], bins=30, color='green', alpha=0.7)
plt.axvline(average_dew_point, color='red', linestyle='dashed', linewidth=2)
plt.title('Dew Point Distribution')
plt.xlabel('Dew Point (F)')
plt.ylabel('Frequency')
plt.legend(['Average'])
```

```python
# Correct column name for wind speed
wind_speed_column_name = 'Wind Speed'

# Wind Speed visualization
plt.hist(data[wind_speed_column_name], bins=30, color='orange', alpha=0.7)
plt.axvline(data[wind_speed_column_name].mean(), color='red', linestyle='dashed', linewidth=2)
plt.title('Wind Speed Distribution')
plt.xlabel('Wind Speed (mph)')
plt.ylabel('Frequency')
plt.legend(['Average'])

plt.show()
```

```python
plt.subplot(2, 2, 4)
correlation_matrix = data[['Temperature', 'Dew Point', 'Wind Speed']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')

plt.tight_layout()
plt.show()
```

```python
# Scatter plot matrix
plt.subplot(2, 2, 4)
sns.pairplot(data[['Temperature', 'Dew Point', 'Wind Speed']], kind='scatter', diag_kind='kde')
plt.title('Scatter Plot Matrix')

plt.tight_layout()
plt.show()
```

```python
# Temperature violin plot
plt.subplot(2, 3, 4)
sns.violinplot(x='Temperature', data=data, color='blue', inner='quartile')
plt.title('Temperature Distribution')
plt.xlabel('Temperature (F)')
plt.ylabel('Density')

plt.tight_layout()
plt.show()

```

```python
# Dew Point violin plot
plt.subplot(2, 3, 5)
sns.violinplot(x='Dew Point', data=data, color='green', inner='quartile')
plt.title('Dew Point Distribution')
plt.xlabel('Dew Point (F)')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
# Wind Speed violin plot
plt.subplot(2, 3, 6)
sns.violinplot(x='Wind Speed', data=data, color='orange', inner='quartile')
plt.title('Wind Speed Distribution')
plt.xlabel('Wind Speed (mph)')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```
