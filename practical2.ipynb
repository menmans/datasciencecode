```python
avg_1st_marks=df['1st'].astype('float64').mean(axis=0)
avg_2nd_marks=df['2nd'].astype('float64').mean(axis=0)
avg_3rd_marks=df['3rd'].astype('float64').mean(axis=0)
avg_4th_marks=df['4th'].astype('float64').mean(axis=0)
avg_5th_marks=df['5th'].astype('float64').mean(axis=0)
```

```python
print("Average Marks of 1st: ",avg_1st_marks)
print("Average Marks of 2nd: ",avg_2nd_marks)
print("Average Marks of 3rd: ",avg_3rd_marks)
print("Average Marks of 4th: ",avg_4th_marks)
print("Average Marks of 5th: ",avg_5th_marks)
```

```python
df['1st'].replace(np.nan , avg_1st_marks , inplace=True)
df['2nd'].replace(np.nan , avg_2nd_marks , inplace=True)
df['3rd'].replace(np.nan , avg_3rd_marks , inplace=True)
df['4th'].replace(np.nan , avg_4th_marks , inplace=True)
df['5th'].replace(np.nan , avg_5th_marks , inplace=True)
```

```python
max_1st=df['1st'].max()
max_2nd=df['2nd'].max()
max_3rd=df['3rd'].max()
max_4th=df['4th'].max()
max_5th=df['5th'].max()
```

```python
print('Max 1st:',max_1st)
print('Max 2nd:',max_2nd)
print('Max 3rd:',max_3rd)
print('Max 4th:',max_4th)
print('Max 5th:',max_5th)
```

```python
cgpa_col=['1st','2nd','3rd','4th','5th']
```

```python
max_values=[max_1st,max_2nd,max_3rd,max_4th,max_5th]
```

```python
for col,max_val in zip(cgpa_col,max_values):
  df[col+'_percentage']=(df[col]/max_val)*100
```
